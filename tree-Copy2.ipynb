```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import load_diabetes
```

```python
import pandas as pd
diabetes = load_diabetes()
df = pd.DataFrame(diabetes["data"],columns=diabetes["feature_names"])
df["target"] = diabetes["target"]
df
```

```python
# 切開來 -> 為之後步驟做準備
from sklearn.model_selection import train_test_split
# [90%題目train,10%test, 90%答案, 10%test答案]
x_train, x_test, y_train, y_test = train_test_split(diabetes["data"], diabetes["target"], test_size=0.1)
```

```python
# (90%)算出方程式 -> fit
reg = DecisionTreeRegressor(max_depth=5)
reg.fit(x_train, y_train)
```

```python
import graphviz
from sklearn.tree import export_graphviz
g = export_graphviz(reg, feature_names=diabetes["feature_names"], 
                filled=True)
graphviz.Source(g)
```

```python
from sklearn.metrics import accuracy_score
# (10%)驗證方程式好不好
pre = reg.predict(x_test)
print("預測：", list(pre))
print("正確：", list(y_test))
```
